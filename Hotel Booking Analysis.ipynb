# Data Analysis And Visualization

```python
# Calculate the percentage of canceled vs. not canceled reservations
reservation_perc = df['is_canceled'].value_counts(normalize = True)
print(reservation_perc)

# Create a bar chart to visualize the reservation status counts
plt.figure(figsize = (5,4))  # Set figure size to 5x4 inches
plt.title('Reservation Status Count')  # Add title to the plot
plt.bar(['Not canceled','Canceled'], df['is_canceled'].value_counts(), edgecolor = 'k', width = 0.8 )  # Create bar chart with black edges
plt.show()  # Display the plot
```

```python
# Create a figure with specified size (8x4 inches)
plt.figure(figsize = (8,4))
# Create a count plot showing hotel types with cancellation status as different colors
ax1 = sns.countplot(x = 'hotel', hue = 'is_canceled', data = df, palette = 'Reds')
# Get the legend handles and labels (though not used later)
legend_labels,_ = ax1.get_legend_handles_labels()
# Position the legend outside the plot at the top-right
ax1.legend(bbox_to_anchor=(1,1))
# Set the title of the plot with specified font size
plt.title('Reservation status in differnet hotels', size = 20)
# Label the x-axis
plt.xlabel('hotel')
# Label the y-axis
plt.ylabel('number of reservations')
# Override the legend with custom labels
plt.legend(['not canceled', 'canceled'])
# Display the plot
plt.show()
```

```python
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize = True)
```

```python
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize = True)
```

```python
# Group the resort_hotel dataframe by reservation_status_date and calculate the mean ADR (Average Daily Rate) for each date
resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()

# Group the city_hotel dataframe by reservation_status_date and calculate the mean ADR (Average Daily Rate) for each date
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
# Plot the ADR trends
plt.figure(figsize=(12,6))  # Set chart size

plt.plot(resort_hotel.index, resort_hotel['adr'], label='Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label='City Hotel')

# Add chart title and labels
plt.title('Average Daily Rate (ADR) Over Time', fontsize=14)
plt.xlabel('Reservation Date', fontsize=12)
plt.ylabel('Average Daily Rate (ADR)', fontsize=12)

# Add legend and grid for better readability
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Extract month from reservation_status_date and create a new column
df['month'] = df['reservation_status_date'].dt.month

# Create a figure with specified size
plt.figure(figsize = (16,8))

# Create a countplot showing cancellation distribution by month
ax1 = sns.countplot(x = 'month', hue = 'is_canceled', data = df, palette = 'bright')

# Get legend handles and labels
legend_labels,_ = ax1.get_legend_handles_labels()

# Position the legend outside the plot
ax1.legend(bbox_to_anchor=(1,1))

# Add title and axis labels
plt.title('Reservation status per month', size = 20)
plt.xlabel('month')
plt.ylabel('number of reservations')

# Override the default legend with custom labels
plt.legend(['not canceled', 'canceled'])

# Display the plot
plt.show()
```

```python
plt.figure(figsize=(15,8))  # Create a new figure with specified size (width=15, height=8 inches)
plt.title('ADR per Month', fontsize=30)  # Set the title with large font size
# Create the bar plot with different colors
ax = sns.barplot(
    x='month',  # X-axis represents months
    y='adr',    # Y-axis represents Average Daily Rate
    data=df[df['is_canceled'] == 1].groupby('month')[['adr']].sum().reset_index(),  # Filter for canceled bookings, group by month, sum ADR values
    palette='pastel',  # Use pastel color scheme for bars
    edgecolor='black'  # Add black outline to bars
)
# Add value labels on top of each bar
for i in ax.containers:
    ax.bar_label(i, fmt='%.1f', fontsize=12)  # Format labels to 1 decimal place with font size 12
plt.show()  # Display the plot
```

```python
def plot_average_daily_rate_by_cancellation(df: pd.DataFrame, start_year: int = 2016, end_year: int = 2017) -> None:
    """
    Creates a line plot showing the average daily rate for canceled and non-canceled bookings
    for the specified year range, with months properly ordered chronologically.
    
    Args:
        df (pd.DataFrame): DataFrame containing hotel booking data with 'arrival_date_year',
                          'arrival_date_month', 'is_canceled', and 'adr' (average daily rate) columns
        start_year (int): Starting year for the analysis (inclusive)
        end_year (int): Ending year for the analysis (inclusive)
    
    Returns:
        None: Displays the plot
    """
    # Filter data for the specified years
    filtered_df = df[(df['arrival_date_year'] >= start_year) & 
                     (df['arrival_date_year'] <= end_year)].copy()
    
    # Month name to number mapping for proper chronological ordering
    month_order = {
        'January': '01', 'February': '02', 'March': '03', 'April': '04',
        'May': '05', 'June': '06', 'July': '07', 'August': '08',
        'September': '09', 'October': '10', 'November': '11', 'December': '12'
    }
    
    # Create a proper datetime-sortable yearmonth column
    filtered_df['yearmonth_sort'] = filtered_df['arrival_date_year'].astype(str) + '-' + \
                                   filtered_df['arrival_date_month'].map(month_order)
    
    # Create a display version for the x-axis labels
    filtered_df['yearmonth_display'] = filtered_df['arrival_date_year'].astype(str) + '-' + \
                                      filtered_df['arrival_date_month']
    
    # Calculate average daily rate by cancellation status
    adr_by_status = filtered_df.groupby(['yearmonth_sort', 'yearmonth_display', 'is_canceled'])['adr'].mean().reset_index()
    
    # Pivot the data for plotting
    pivot_df = adr_by_status.pivot_table(
        index=['yearmonth_sort', 'yearmonth_display'], 
        columns='is_canceled', 
        values='adr'
    ).reset_index()
    
    # Sort chronologically
    pivot_df = pivot_df.sort_values('yearmonth_sort')
    
    # Set more descriptive column names
    pivot_df = pivot_df.rename(columns={0: 'Not Canceled', 1: 'Canceled'})
    
    # Create and configure the plot
    plt.figure(figsize=(12, 6))
    
    # Plot using the sorted data
    plt.plot(pivot_df['yearmonth_display'], pivot_df['Not Canceled'], marker='o', label='Not Canceled')
    plt.plot(pivot_df['yearmonth_display'], pivot_df['Canceled'], marker='o', label='Canceled')
    
    # Set plot attributes
    plt.title(f'Average Daily Rate by Cancellation Status ({start_year}-{end_year})')
    plt.xlabel('Year-Month')
    plt.ylabel('Average Daily Rate')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    
    # Show the plot
    plt.show()
```

```python
# Assuming 'hotel_df' is your DataFrame with the booking data
import matplotlib.pyplot as plt
plot_average_daily_rate_by_cancellation(df)
```
